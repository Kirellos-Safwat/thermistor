# **Thermistor**

### $Temperature(°{C}) - Resistance({Ω}) curve$

```python
import numpy as np
import matplotlib.pyplot as plt

# Steinhart-Hart coefficients for the thermistor
A = 1.009249522e-03
B = 2.378405444e-04
C = 2.019202697e-07

# Function to calculate temperature from resistance
def resistance_to_temperature(R):
    return 1 / (A + B * np.log(R) + C * (np.log(R))**3) - 273.15  # Convert Kelvin to Celsius

# Range of resistance values
R_values = np.linspace(100, 10000, 100)  # Example resistance range from 100 to 10,000 Ω

# Calculate corresponding temperatures
temperatures = resistance_to_temperature(R_values)

# Plot resistance-temperature curve
plt.plot(temperatures, R_values)
plt.title('Resistance-Temperature Relationship of Thermistor')
plt.ylabel('Resistance (Ω)')
plt.xlabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

### **Calibration Curve**

```python
import numpy as np
import matplotlib.pyplot as plt

# Provided data points
data = np.array([
    [32.5, 7.87],
    [33.6, 7.82],
    [34.8, 7.78],
    [35.5, 7.74],
    [39.4, 7.68],
    [40.2, 7.64],
    [41.5, 7.5],
    [42.7, 7.34]
])

# Splitting data into x and y arrays
x = data[:, 0]
y = data[:, 1]

# Perform linear regression (fitting a first-degree polynomial)
coefficients = np.polyfit(x, y, 1)

# Get the slope and intercept of the best-fit line
slope = coefficients[0]
intercept = coefficients[1]

# Generate points for the best-fit line
best_fit_x = np.linspace(min(x), max(x), 100)
best_fit_y = slope * best_fit_x + intercept

# Plot the original data points
plt.scatter(x, y, label='Data')

# Plot the best-fit line
plt.plot(best_fit_x, best_fit_y, color='red', label='Best Fit Line')

# Add labels and legend
plt.xlabel('Temperature (°C)')
plt.ylabel('Voltage (V)')
plt.title('Best Fit Line')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
